```python
# The file with the TSP problem
filename = 'wi29.tsp'
```

```python
###
# Based on:
# https://github.com/perrygeo/simanneal and its examples
###

# Main function and classes, the algorithm itself

import math
import random

from simanneal import Annealer

import matplotlib.animation as animation
import matplotlib.pyplot as plt
import seaborn as sns


def distance(x, y):
    """
    Calculates the Eclidean 2D distance between two points
    """
    return math.sqrt(((x[0] - y[0]) ** 2) + ((x[1] - y[1]) ** 2))


def read_cities(filename):
    """
    Reads a file with a TSP instance
    """
    cities = {}
    reading = False
    with open(filename, 'r') as tsp_file:
        for line in tsp_file:
            if not reading:
                if line.startswith('1'):
                    reading = True
                else:
                    continue
            else:
                if line.startswith('EOF'):
                    continue

            splited = line.split()
            cities[int(splited[0])] = [float(splited[1]), float(splited[2])]
    
    return cities


def create_distance_matrix(cities):
    """
    Creates the distance matrix
    """
    distance_matrix = {}
    for ka, va in cities.items():
        distance_matrix[ka] = {}
        
        for kb, vb in cities.items():
            if kb == ka:
                distance_matrix[ka][kb] = 0.0
            else:
                distance_matrix[ka][kb] = distance(va, vb)
                
    return distance_matrix


class TravellingSalesmanProblem(Annealer):
    """
    Test annealer with a travelling salesman problem
    """
    
    def __init__(self, state, distance_matrix):
        """
        Pass extra data (the distance matrix) into the constructor
        """
        self.distance_matrix = distance_matrix
        super(TravellingSalesmanProblem, self).__init__(state)

    def move(self):
        """
        Swaps two cities in the route
        """
        a = random.randint(0, len(self.state) - 1)
        b = random.randint(0, len(self.state) - 1)
        self.state[a], self.state[b] = self.state[b], self.state[a]

    def energy(self):
        """
        Calculates the length of the route
        """
        e = 0
        for i in range(len(self.state)):
            e += self.distance_matrix[self.state[i-1]][self.state[i]]
        return e
```

```python
# The data for the algorithm, its params and the running

# Cities, initial state (a randomly-ordered itinerary of the cities) and the distance matrix
cities = read_cities('tsp_files/' + filename)
init_state = list(cities.keys())
random.shuffle(init_state)
distance_matrix = create_distance_matrix(cities)

tsp = TravellingSalesmanProblem(init_state, distance_matrix)
tsp.copy_strategy = "slice"  # Since our state is just a list, slice is the fastest way to copy

# An optional config
# auto_schedule = tsp.auto(minutes=0.5)  #  Run alg. for 30 secs
# tsp.set_schedule(auto_schedule)

best_state, distance, best_states_list = tsp.anneal()

print "Best route: %i units:" % distance
print best_state
```

```python
# Plotting best solution

fig, ax = plt.subplots()
ax.plot([cities[state][0] for state in best_state], [cities[state][1] for state in best_state], linestyle='-', marker='o')
```

```python
###
# Ploting from:
# http://nbviewer.ipython.org/github/jrjohansson/scientific-python-lectures/blob/master/Lecture-4-Matplotlib.ipynb
###

# Plotting all solutions in `.mp4`

fig, ax = plt.subplots()

ax.set_xlim([20000, 28000])
ax.set_ylim([9000, 18000])

path, = ax.plot([], [], linestyle='-', marker='o')

def init():
    path.set_data([], [])
    
def update(n):
    path.set_data([cities[state][0] for state in best_states_list[n]], [cities[state][1] for state in best_states_list[n]])

anim = animation.FuncAnimation(fig, update, init_func=init, frames=len(best_states_list), blit=True)

anim.save('tsp_' + filename + '_simulated_annealing.mp4', fps=20, writer="avconv", codec="libx264")

plt.close(fig)
```
